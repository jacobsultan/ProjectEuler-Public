Triangle, pentagonal, and hexagonal numbers are generated by the following formulae:
<table><tr><td>Triangle</td>
<td> </td>
<td>$T_n=n(n+1)/2$</td>
<td> </td>
<td>$1, 3, 6, 10, 15, \dots$</td>
</tr><tr><td>Pentagonal</td>
<td> </td>
<td>$P_n=n(3n - 1)/2$</td>
<td> </td>
<td>$1, 5, 12, 22, 35, \dots$</td>
</tr><tr><td>Hexagonal</td>
<td> </td>
<td>$H_n=n(2n - 1)$</td>
<td> </td>
<td>$1, 6, 15, 28, 45, \dots$</td>
</tr></table>It can be verified that $T_{285} = P_{165} = H_{143} = 40755$.
Find the next triangle number that is also pentagonal and hexagonal.


```python
import time
start_time = time.time()
```

```python
notfound = []

trii = 2
penti = 2
hexi = 2

tris = [1]
pents = [1]
hexs = [1]
while len(notfound) < 2:
    
    while pents[-1]>= tris[-1] <=hexs[-1]:
        next = int(trii * (trii + 1) / 2)
        if next in pents and next in hexs and next not in notfound:
            print(next)
            notfound.append(next)
            break
        tris.append(int(trii * (trii + 1) / 2))
        trii += 1
        
    while tris[-1] >= pents[-1] <= hexs[-1]:
        next = int(penti * (3 * penti - 1) / 2)
        if next in hexs and next in tris and next not in notfound:
            print(next)
            notfound.append(next)
            break
        pents.append(next)
        penti += 1
        
    while tris[-1] >= hexs[-1] <= pents[-1]:
        next = int(hexi * (2 * hexi - 1))
        if next in tris and next in pents and next not in notfound:
            print(next)
            notfound.append(next)
            break
        hexs.append(int(hexi * (2 * hexi - 1)))
        hexi += 1
print(notfound)
        
    
```

```python
print("Process finished --- %s seconds ---" % (time.time() - start_time))
```
